```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
```

## Step 1: Read and clean the data

```python
true_news = pd.read_csv("data/True.csv")
fake_news = pd.read_csv("data/Fake.csv")

pd.concat([true_news.head(), true_news.tail()])
```

```python
pd.concat([fake_news.head(), fake_news.tail()])
```

Concatenate the real and fake datasets, reset the index, and shuffle the combined data to avoid order bias

```python
fake_news["label"] = 0
true_news["label"] = 1

news = pd.concat([fake_news, true_news], ignore_index=True)
news = news.sample(frac=1, random_state=42).reset_index(drop=True)

news.head()
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text_without_punctuation = re.sub(r'[^\w\s]', '', text)
    text_without_digits = re.sub(r'\d+', '', text_without_punctuation)
    text_lc = text_without_digits.lower()
    tokens = word_tokenize(text_lc)

    clean_text = ' '.join([word for word in tokens if word not in stop_words])

    return clean_text


news["text"] = news["text"].apply(clean_text)

news.head()
```

```python
news = news[["text", "label"]]

news.to_csv("data/news.csv", index=0)
```
